```python
def geomark(label_dir,geoplots_dir,data,gfp):
    for filename in os.listdir(label_dir):
        if filename.endswith('.csv'):
            filepath = os.path.join(label_dir,filename)
        labels = pd.read_csv(filepath)
        data['Label'] = labels['Label']
        merge = pd.merge(gfp,data,left_on='id',right_on='Alpha-3')
        geoplot = merge[['id','name','Label','geometry']]
        savepath = os.path.join(geoplots_dir,filename)
        geoplot.to_csv(savepath,index=False)
```

```python
label_dir = r'E:\Projects\National Anthems\labels'
geoplot_dir = r'E:\Projects\National Anthems\data\geoplots'
data = pd.read_csv(r'E:\Projects\National Anthems\data\national_anthems.csv')
```

```python
geomark(label_dir,geoplot_dir,data,gfp)
```
