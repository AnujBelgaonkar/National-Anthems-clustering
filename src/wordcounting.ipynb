```python
import pandas as pd
from collections import Counter
from nltk import word_tokenize
import os
```

```python
def processtext(text):
    tokens = word_tokenize(text)
    return tokens
```

```python
def wordcount(df):
    df['tokens'] = df['Proessed_Anthem'].apply(processtext)
    category_word_count = {}
    for index,row in df.iterrows():
        category = row['Label']
        tokens = row['tokens']
        if category not in category_word_count:
            category_word_count[category] = Counter()
        category_word_count[category].update(tokens)
    word_count_df = pd.DataFrame()
    for category,word_count in category_word_count.items():
        top_words = word_count.most_common(20)
        category_df = pd.DataFrame(top_words,columns=['word','count'])
        category_df['category'] = category
        word_count_df = pd.concat([word_count_df,category_df])
    word_count_df.reset_index(drop=True,inplace=True)
    return word_count_df
```

```python
def word_count_each(data,in_directory,out_directory):
    for filename in os.listdir(in_directory):
        if filename.endswith('.csv'):
            filepath = os.path.join(in_directory,filename)
        df = pd.read_csv(filepath)
        data['Label'] = df['Label']
        word_count_df = wordcount(data)
        output_filename = f'word_count_df_{filename}'
        output_path = os.path.join(out_directory,output_filename)
        word_count_df.to_csv(output_path,index = False)
```

```python
in_directory = r"E:\Projects\National Anthems\labels"
out_directory = r"E:\Projects\National Anthems\data\wordcounts"
```
