```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
data = pd.read_csv(r'E:\Projects\National Anthems\data\vectors.csv')
```

```python
data
```

```python
wcc = []
x = []
for i in range(2,20):
    kmeans = KMeans(n_clusters=i).fit(data)
    wcc.append(kmeans.inertia_)
    x.append(i)
```

```python
plt.plot(x,wcc)
```

```python
kmeans.labels_
```

```python
for i in range(5,21):
    kmeans = KMeans(n_clusters=i).fit(data)
    t_df = pd.DataFrame(kmeans.labels_)
    t_df.columns = ['Label']
    t_df.to_csv(f"E:/Projects/National Anthems/labels/kmeans{i}",index=False)
```
