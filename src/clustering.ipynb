```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'E:\Projects\National Anthems\data\vectors.csv')
```

```python
data
```

```python
wcc = []
x = []
for i in range(2,20):
    kmeans = KMeans(n_clusters=i).fit(data)
    wcc.append(kmeans.inertia_)
    x.append(i)
```

```python
plt.plot(x,wcc)
```

```python
kmeans.labels_
```

```python
for i in range(5,21):
    kmeans = KMeans(n_clusters=i).fit(data)
    t_df_labels = pd.DataFrame(kmeans.labels_)
    t_df_centroids = pd.DataFrame(kmeans.cluster_centers_)
    t_df_labels.columns = ['Label']
    t_df_labels.to_csv(f"E:/Projects/National Anthems/labels/kmeans{i}.csv",index=False)
    t_df_centroids.to_csv(f'E:/Projects/National Anthems/lists/clustercenters/kmeans{i}.csv',index=False)
```
